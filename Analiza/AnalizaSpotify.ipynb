Wyrzucam kolumny, które nie będą mieć znaczenia dla tej analizy tj. episode_name, episode_show_name, spotify_episode_uri.

```python
dfa = dfa.drop(['episode_name', 'episode_show_name', 'spotify_episode_uri'], axis = 1)
```

```python
dfa
```

```python
sns.countplot(data = dfa, y = 'platform', order = dfa['platform'].value_counts().index)
```

```python
dfp = dfa['master_metadata_track_name'].value_counts()
dfp = dfp[dfp > 96]
dfp = pd.DataFrame(dfp)
```

```python
dfp
```

```python
sns.barplot(data = dfp,y ='master_metadata_track_name', x = 'count')
plt.xlabel('Liczba odsłuchań')
plt.ylabel('Tytuł piosenki')
plt.title('Najczęściej słuchane piosenki > 95')
```

```python
dft = dfa['ts'].str[:-10]
```

```python
dfa['ts'] = dft
```

```python
dfa['ts'] = pd.to_datetime(dfa['ts'])
```

```python
dft1 = dfa.groupby('ts')['ms_played'].sum()
```

```python
dft1 = pd.DataFrame(dft1)
```

```python
dft1['ms_played'] = (dft1['ms_played'] / 60000).round()
```

```python
dft1 = dft1.reset_index()
```

```python
dft1.info()
```

```python
df_grouped = dft1.groupby(dft1['ts'].dt.year)['ms_played'].sum().reset_index()
```

```python
sns.lineplot(data = df_grouped, x = 'ts', y = 'ms_played')
plt.xlabel('Czas słuchania w minutach')
plt.ylabel('Rok')
plt.title('Rok a czas spędzony na słuhaniu muzyki')
```
